# Plotting the data of median less than 20

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob,os

lcs_data =glob.glob('lcs/*.csv')
print(len(lcs_data))
arr =[]
val =20
count =0
c =0
n_bins =20

for i in lcs_data:
     df =pd.read_csv(i)
     df.columns =['mjd','mag','err']
     path=os.path.splitext(i)
     med =np.median(df['mag'])
     arr.append(med)
    
       
# Creating histogram
fig, axs = plt.subplots(1, 1,
                        figsize =(10, 7),
                        tight_layout = True)
axs.hist(arr, bins = n_bins,histtype='bar',rwidth=0.8)

# Adding extra features   
plt.xlabel("median of magnitude in lightcurves")
plt.ylabel("number of lightcurves")
plt.title('median of magnitude in lightcurves v/s number of lightcurves in histogram')

# Show plot
plt.show()
```
