```python
import numpy as np
import pandas as pd
from ztfquery import lightcurve
import csv
#ra,dec=np.genfromtxt("raw_data.csv",unpack=True,usecols=(0,1))
data=pd.read_csv('raw_data.csv',usecols =(0,1))
ra=data['RA']
dec =data['DEC']
username=input('Enter Username for IRSA:\t')
password=input('Enter Password for IRSA:\t')
count=0
val =20
for i in range(168295,200000):
    
    print(ra[i],dec[i])
    data = lightcurve.LCQuery.download_data(circle=[ra[i],dec[i],0.0014], bandname="g",auth=[username,password])
    date_init=data['mjd']
    mag_init=data['mag']
    magerr_init=data['magerr']
    catflag=data['catflags']
    ra_init=data['ra']
    dec_init=data['dec']
    #good_id = np.where(catflag == 0)[0]
    #date,mag,magerr,ra,dec= date[good_id],mag[good_id],magerr[good_id],ra[good_id],dec[good_id]
    #Date.append(date)
    #Mag.append(mag)
    #Mag_Err.append(magerr)
    #RA.append(ra)
    #DEC.append(dec)
    def check(mag_init, val):
        return(all(x >= val for x in mag_init))
    if len(date_init)==0:
        print('file has no data points'.format(i+1))
        with open("./missing/{}_{}.g.csv".format(ra[i],dec[i]), 'w+') as csvfile:
            csvwriter = csv.writer(csvfile)
            print("Light curve no {}.".format(i+1))
        count+=1
    elif len(date_init)<=100:
        print('less than 100 data points'.format(i+1))
        with open("./less/{}_{}.g.csv".format(ra[i],dec[i]), 'w+') as csvfile:
            csvwriter = csv.writer(csvfile)
            print("Light curve no {}.".format(i+1))
        count+=1
    elif (check(mag_init,val)):
        print('magnitude is greater than or equal to 20'.format(i+1))
        with open("./greater/{}_{}.g.csv".format(ra[i],dec[i]), 'w+') as csvfile:
            csvwriter = csv.writer(csvfile)
            print("Light curve no {}.".format(i+1))
        count+=1
    else:
        with open("./lcs/{}_{}.g.csv".format(ra[i],dec[i]), 'w+') as csvfile:
            csvwriter = csv.writer(csvfile)
            csvwriter.writerows(zip(date_init,mag_init,magerr_init))
            print("Light curve no {} downloaded.".format(i+1))
print(count)
```
